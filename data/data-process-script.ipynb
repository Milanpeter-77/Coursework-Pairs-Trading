```python
import pandas as pd

df = pd.read_csv('DJGT_raw.csv')

# drop permno
df = df.drop(columns=['PERMNO'])

# rename date to DATE
df = df.rename(columns={'date': 'DATE'})

# pivot the dataframe to have one row per date and columns for each variable
df_pivot = df.pivot(index='DATE', columns='TICKER', values='PRC')

# export the pivoted dataframe to a new CSV file
df_pivot.to_csv('DJGT_prices.csv')
```
